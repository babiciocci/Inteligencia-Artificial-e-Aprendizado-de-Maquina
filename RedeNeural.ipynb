```python
m_unico=(Simulated_DF['m (kg)'].unique())
M_unico=(Experimental_DF['m (kg)'].unique())
```

```python
classificacoes = ['A', 'B', 'C', 'D', 'E', 'F']

mapeamento = dict(zip(m_unico, classificacoes))

Simulated_DF['classificacao'] = Simulated_DF['m (kg)'].map(mapeamento)
Experimental_DF['classificacao'] = Experimental_DF['m (kg)'].map(mapeamento)
```

```python
print(f"Quantidade de valores diferentes de m (experimental): {Experimental_DF['m (kg)'].nunique()}")
print(f"Quantidade de valores diferentes de m (simulado): {Simulated_DF['m (kg)'].nunique()}")
```

```python
Experimental_DF.drop(columns=['aceleracao calc (m/s^2)', 'pctg'], inplace=True)
Experimental_DF.rename(columns={'aceleracao exp (m/s^2)': 'aceleracao (m/s^2)'}, inplace=True)

Experimental_DF
```

```python
x_train = Simulated_DF[['M (kg)', 'aceleracao (m/s^2)']]
x_test = Experimental_DF[['M (kg)', 'aceleracao (m/s^2)']]
```

```python
y_train = Simulated_DF['classificacao']
y_test = Experimental_DF['classificacao']

y_train_regressor = Simulated_DF['m (kg)']
y_test_regressor = Experimental_DF['m (kg)']
```

```python
modelo_neural = MLPClassifier(hidden_layer_sizes=(2, 3), max_iter=6000,
                              random_state=42, activation = 'relu')

# Treinar o modelo usando os dados simulados
modelo_neural.fit(x_train, y_train)

# Avaliar a acurácia nos dados de teste utilizando os
# dados experimentais
acuracia_neural = modelo_neural.score(x_test, y_test)

print(f"A acurácia da rede neural é: {acuracia_neural * 100:.2f}%")
```
